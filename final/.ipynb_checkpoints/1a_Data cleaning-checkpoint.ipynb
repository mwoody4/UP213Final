```python
import requests

url = (
    "https://api.census.gov/data/2021/acs/acs5?"
    "get=B22010_001E,B22010_002E,B17017_002E,B17017_006E,NAME"
    "&for=county:037&in=state:06"
)

response = requests.get(url)
data = response.json()

for row in data:
    print(row)
```

```python
# Parse data
header, values = data[0], data[1]

# Convert strings to integers
B22010_001E = int(values[0])  # Total households
B22010_002E = int(values[1])  # SNAP households
B17017_002E = int(values[2])  # Poverty households
B17017_006E = int(values[3])  # Poverty, no SNAP

# Calculate rates
observed_snap_rate = (B22010_002E / B22010_001E) * 100 if B22010_001E > 0 else 0
ebne_rate = (B17017_006E / B17017_002E) * 100 if B17017_002E > 0 else 0

# Residual SNAP gap
snap_gap = observed_snap_rate - ebne_rate

# Output
print(f"Observed SNAP Participation Rate: {observed_snap_rate:.2f}%")
print(f"EBNE Rate: {ebne_rate:.2f}%")
print(f"Residual SNAP Gap: {snap_gap:.2f}%")
```
